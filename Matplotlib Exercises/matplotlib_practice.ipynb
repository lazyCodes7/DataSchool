```python
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4,5,6,7]
y=[0,1,4,9,16,25,36,49]
```

```python
fig, axes = plt.subplots(figsize=(8,3),nrows=1,ncols=2)
for current_axis in axes:
    current_axis.plot(x,y)
plt.tight_layout()
```

```python
fig= plt.figure(figsize=(7,3))
axes = fig.add_axes([0,0,0.7,1])
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.plot(x,y,color='purple',linewidth=2,linestyle=':',marker='o',markersize=10) 
axes.set_xlim([0,1]) # For plotting graph between a specific range
axes.set_ylim([0,1])
# Various params in matplotlib library for plotting 
# linewidth is for specifying width of the line
# linestyle is for styling line appearance
# marker points the data points used for plotting the graph
# markersize for specifying size
# alpha for transparency
```

```python
fig.savefig('my_pictures.png')
```
